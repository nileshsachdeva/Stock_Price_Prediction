13. Enterprise multiple

Overall, this measurement allows investors to assess a company on the same basis as that of an acquirer. As a rough calculation, enterprise value multiple serves as a proxy for how long it would take for an acquisition to earn enough to pay off its costs in years(assuming no change in EBITDA).

```python
def enterprise_multiple(company):
  print("Enterprise value: "+str(company.info['enterpriseValue']))
  print("EBITDA: "+str(company.info['ebitda']))
  print("Enterprise value/EBITDA: "+str(company.info['enterpriseValue']/company.info['ebitda']))
  return company.info['enterpriseValue'],company.info['ebitda'],company.info['enterpriseValue']/company.info['ebitda']
```

15. Long term debt to total assets

```python
# show balance sheet
def longTermDebt_to_TotalAssetss(company):
  print("Annual")
  annual_balance_sheet = company.balance_sheet
  print(annual_balance_sheet.loc['Total Assets'][0:2])
  print(annual_balance_sheet.loc['Long Term Debt'][0:2])
  annbalsheet = pd.DataFrame()
  annbalsheet['Long Term Debt'] = annual_balance_sheet.loc['Long Term Debt'][0:2]
  annbalsheet['Total Assets'] = annual_balance_sheet.loc['Total Assets'][0:2]
  annbalsheet['Date'] = (annual_balance_sheet.loc['Long Term Debt'][0:2].index).strftime("%d-%m-%Y")
  annbalsheet = (annbalsheet).to_json(orient='index')
  print(annbalsheet)
  print()
  print("Long term debt to total assets: "+str(annual_balance_sheet.loc['Long Term Debt'][0]/annual_balance_sheet.loc['Total Assets'][0]))
  annual_longtermdebt_to_totalassets_ratio = annual_balance_sheet.loc['Long Term Debt'][0]/annual_balance_sheet.loc['Total Assets'][0]

  print("Quartely")
  quartely_balance_sheet = company.quarterly_balance_sheet
  print(quartely_balance_sheet.loc['Total Assets'][0:2])
  print(quartely_balance_sheet.loc['Long Term Debt'][0:2])
  quarterbalsheet = pd.DataFrame()
  quarterbalsheet['Long Term Debt'] = quartely_balance_sheet.loc['Long Term Debt'][0:2]
  quarterbalsheet['Total Assets'] = quartely_balance_sheet.loc['Total Assets'][0:2]
  quarterbalsheet['Date'] = (quartely_balance_sheet.loc['Long Term Debt'][0:2].index).strftime("%d-%m-%Y")
  quarterbalsheet = (quarterbalsheet.reset_index()).to_json(orient='index')
  print(quarterbalsheet)
  print()
  print("Long term debt to total assets: "+str(quartely_balance_sheet.loc['Long Term Debt'][0]/quartely_balance_sheet.loc['Total Assets'][0]))
  quarter_longtermdebt_to_totalassets_ratio = quartely_balance_sheet.loc['Long Term Debt'][0]/quartely_balance_sheet.loc['Total Assets'][0]
  return annbalsheet,annual_longtermdebt_to_totalassets_ratio,quarterbalsheet,quarter_longtermdebt_to_totalassets_ratio
```

24.	Price/earning to growth ratio (PEG)

The PEG ratio is a refinement of the P/E ratio and factors in a stock's estimated earnings growth into its current valuation.The general consensus is that if the PEG ratio indicates a value of 1, this means that the market is correctly valuing (the current P/E ratio) a stock in accordance with the stock's current estimated earnings per share growth. If the PEG ratio is less than 1, this means that EPS growth is potentially able to surpass the market's current valuation. In other words, the stock's price is being undervalued. On the other hand, stocks with high PEG ratios can indicate just the opposite - that the stock is currently overvalued.


```python
def peg_ratio(company):
  print("PEG ratio: "+str(company.info['pegRatio']))
  return company.info['pegRatio']
```
